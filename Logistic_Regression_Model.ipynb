```python
#Import Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
#Create a Data set 
data = {
    "Hours" : [1,2,3,4,5,6,7,8],
    "Passed" :[0,0,0,0,1,1,1,1],
}
df = pd.DataFrame(data)
print(df)
```

```python
#Visualise the Data 
plt.scatter(df['Hours'],df['Passed'])
plt.xlabel('Hours')
plt.ylabel('Passed')
plt.title('Study Hours v/s Exam Outcome')
plt.show()
```

```python
#Train The Model
X = df[['Hours']]
Y = df['Passed']

model = LogisticRegression()
model.fit(X,Y)

#Prediction 
prediction = model.predict([[4.5]])
print("Prediction of 4.5 Hrs:" , prediction )
```

```python
#Plot the Logistic Curve 
x_values = np.linspace(0,11,100)
y_values = model.predict_proba(x_values.reshape(-1,1))[:,1]
plt.scatter(df['Hours'],df['Passed'])
plt.plot(x_values, y_values ,color = 'red')
plt.xlabel('Hours')
plt.ylabel('Passed')
plt.title('Logistic Regression Curve')
plt.show()
```
